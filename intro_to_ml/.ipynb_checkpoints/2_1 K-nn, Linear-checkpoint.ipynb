```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (10, 3))

# fit 메소드는 self 오브젝트를 return. 따라서 객체 생성과 fit 메소드는 한 줄에 쓸 수 있다.
for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title('{} 이웃'.format(n_neighbors))
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
axes[0].legend(loc=3)
```

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)
# stratify: 계층화하다. -> 층화추출

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1, 11)

for n_neighbors in neighbors_settings:
    # 모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # 훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings,training_accuracy, label='training accuracy')
plt.plot(neighbors_settings, test_accuracy, label='test accuracy')
plt.ylabel('acc')
plt.xlabel('n_neighbors')
plt.legend()
```

```python
# underfitting -> optimal -> overfitting
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
# k-nn regression (회귀)
```

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
y_test
```

```python
reg.score(X_test, y_test)
```

```python
print('test set prediction (R^2) : {:.2f}'.format(reg.score(X_test, y_test)))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
line = np.linspace(-3, 3, 1000).reshape(-1,1) # -3과 3 사이에 1000개의 데이터 포인트 생성

for n_neighbors, ax in zip([1, 3, 9], axes):
    #1, 3, 9 이웃을 사용한 예측
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=6)
    
    ax.set_title("{} neighbor's train socre: {:.2f} \n test score: {:.2f}".format(n_neighbors, reg.score(X_train, y_train),
                                                                   reg.score(X_test, y_test)))
    ax.set_xlabel("feature")
    ax.set_ylabel("target")
axes[0].legend(['model prediction', 'train data&target', 'test data&target'], loc='best')
```

```python
# underfit, opt, over
```

### 선형 회귀 Linear regression = 최소제곱법 OLS (Ordinary Least Squares)
가장 간단하고 오래된 회귀용 선형 알고리즘. 예측과 타깃 y 사이의 평균제곱오차(mean squared error)를 최소화하는 파라미터 w, b 찾음. 매개변수가 없는 것이 장점이지만, 따라서 모델의 복잡도를 제어할 방법도 없음.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
```

```python
lr
```

```python
print("<w>", 'lr.coef_(weight, coefficient):', lr.coef_)
```

```python
print("<b>", 'lr.intercept_(bias, offset, intercept):', lr.intercept_)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
# underfit. 1차원 데이터셋에선 모델이 매우 단순하여 overfit 걱정 x. (특성이 많은 고차원 데이터셋에서는 overfit 가능성 높다.)
```

```python
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

```python
print('train set score: {:.2f}'.format(lr.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lr.score(X_test, y_test)))
```

```python
# train acc >>> test acc. 즉 과대적합.
# 복잡도를 제어할 수 있는 모델이 필요함. 기본 선형 회귀 다음으로 보편적인 리지 회귀 사용.
```

### Ridge 회귀

#### L2 규제: 모든 특성이 출력에 주는 영향을 최소한으로 만듦. 즉 w의 모든 원소가 0에 가깝게 되길 원함. 가중치의 절댓값 최소화.
규제(regularization)

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print('train set score: {:.2f}'.format(ridge.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge.score(X_test, y_test)))
```

```python
# alpha 매개변수 (기본값 1.0, 얼마나 규제하여 단순화할지.) 변경해가며 최적화.

ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print('train set score: {:.2f}'.format(ridge10.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge10.score(X_test, y_test)))
```

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print('train set score: {:.2f}'.format(ridge01.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge01.score(X_test, y_test)))
```

```python
ridge001 = Ridge(alpha=0.01).fit(X_train, y_train)
print('train set score: {:.2f}'.format(ridge001.score(X_train, y_train)))
print('test set score: {:.2f}'.format(ridge001.score(X_test, y_test)))
```

```python
ridge0 = Ridge(alpha=0).fit(X_train, y_train)
print('ridge0 train set score: {:.2f}'.format(ridge0.score(X_train, y_train)))
print('ridge0 test set score: {:.2f}'.format(ridge0.score(X_test, y_test)))

print('alpha = 0 즉 규제가 전혀 없으면 선형 회귀와 동일')

print('lr train set score: {:.2f}'.format(lr.score(X_train, y_train)))
print('lr test set score: {:.2f}'.format(lr.score(X_test, y_test)))
```

```python
plt.plot(ridge10.coef_, '^', label='Ridge alpha=10')
plt.plot(ridge.coef_, 's', label='Ridge alpha=1')
plt.plot(ridge01.coef_, 'v', label='Ridge alpha=0.1')

plt.plot(lr.coef_, 'o', label='LinearRegression')
plt.xlabel('coef list')
plt.ylabel('coef value')
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.ylim(-25, 25)
plt.legend()
```

```python
mglearn.plots.plot_ridge_n_samples()
```

```python
# 데이터가 충분하면 규제 항은 덜 중요해짐. 데이터가 계속 많아지면 리지 회귀와 선형 회귀의 성능이 같아질 것.
```

### Lasso
#### L1 규제: 마찬가지로 계수를 0에 가깝게 하려고 하나, L2와 달리 어떤 계수는 실제로 0이 되기도 함. 즉 모델에서 완전히 제외되는 특성 생김. feature selection이 자동으로 발생.
#### 라소는 계수 벡터의 L1 norm을 penalty로 사용. 즉 계수의 절댓값의 합. 평균제곱오차식에 alpha*sigma(  | wj |  ) 항 추가됨.
Elastic-Net 방식 (L1, L2 규제 함께 사용) 에서 L2 규제 빠진 걸로도 볼 수 있음

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print('train set score: {:.2f}'.format(lasso.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso.score(X_test, y_test)))
print('used feature n:', np.sum(lasso.coef_ != 0))
```

```python
print('Ridge used feature n:', np.sum(ridge.coef_ != 0))
```

```python
# 104 개 중 4개만 씀...
```

```python
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print('train set score: {:.2f}'.format(lasso001.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso001.score(X_test, y_test)))
print('used feature n:', np.sum(lasso001.coef_ != 0))
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print('train set score: {:.2f}'.format(lasso00001.score(X_train, y_train)))
print('test set score: {:.2f}'.format(lasso00001.score(X_test, y_test)))
print('used feature n:', np.sum(lasso00001.coef_ != 0))
```

```python
# alpha 값을 너무 낮추면 규제 효과 x -> 과대적합 -> linear regression과 다를 바 없음
```

```python
plt.plot(lasso.coef_, 's', label = 'lasso alpha=1')
plt.plot(lasso001.coef_, '^', label = 'lasso alpha = 0.01')
plt.plot(lasso00001.coef_, 'v', label = 'lasso alpha = 0.0001')

plt.plot(ridge01.coef_, 'o', label = 'ridge alpha = 0.1')
plt.legend(ncol=2, loc=(0,1.2))
plt.ylim=(-25,25)
plt.xlabel('coefs')
plt.ylabel('coef val')
```

#### Linear model - classification (1. logistic regression 2. support vector machine)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10,3))

for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=.7)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
axes[0].legend()
```

```python
# LogisticRegression과 LinearSVC에서 규제의 강도를 결정하는 매개변수는 C. C의 값이 높아지면 규제 감소. (C 너무 크면 과대적합)
```

### stratify
"stratify" parameter makes a split so that the proportion of values in the sample produced will be the same as the proportion of values provided to parameter stratify.

For example, if variable y is a binary categorical variable with values 0 and 1 and there are 25% of zeros and 75% of ones, stratify=y will make sure that your random split has 25% of 0's and 75% of 1's.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=42)
logreg = LogisticRegression().fit(X_train, y_train)
print('train set score: {:.3f}'.format(logreg.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg.score(X_test, y_test)))
print('used feature n:', np.sum(logreg.coef_ != 0))
```

```python
np.count_nonzero(cancer.feature_names)
```

```python
logreg100 = LogisticRegression(C=100).fit(X_train, y_train)
print('train set score: {:.3f}'.format(logreg100.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg100.score(X_test, y_test)))
```

```python
logreg001 = LogisticRegression(C=0.01).fit(X_train, y_train)
print('train set score: {:.3f}'.format(logreg001.score(X_train, y_train)))
print('test set score: {:.3f}'.format(logreg001.score(X_test, y_test)))
```

```python
plt.plot(logreg.coef_.T) # .T = transpose
```

```python
plt.plot(logreg100.coef_.T, 's', label = 'C=100')
plt.plot(logreg.coef_.T, '^', label = 'C=1(default)')
plt.plot(logreg001.coef_.T, 'v', label = 'C=0.01')

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim=(-5,5)
plt.xlabel('feature')
plt.ylabel('coef val')
plt.legend()
```

```python
# Logistic Regression은 기본으로 L2 규제 적용. Ridge와 비슷한 모양.
```

```python
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(C=C, penalty='l1').fit(X_train, y_train)
    print('C={:.3f}인 L1 로지스틱 회귀 train accuracy: {:.3f}'.format(C, lr_l1.score(X_train, y_train)))
    print('C={:.3f}인 L1 로지스틱 회귀 test accuracy: {:.3f}'.format(C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label='C={:.3f}'.format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim=(-5,5)
plt.xlabel('feature')
plt.ylabel('coef val')
plt.legend()
```
